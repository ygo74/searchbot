# Initialize the client

```python
from qdrant_client import QdrantClient

client = QdrantClient("localhost", port=6333)
```

## Performing queries

```python
# FastEmbed
# embeddings = FastEmbedEmbeddings()

# OpenAIEmbeddings
from dotenv import load_dotenv
import os
import sys

load_dotenv()
base_url=os.getenv('AZURE_OPENAI_ENDPOINT')
print(f"API URL : {base_url}")

from openai import AzureOpenAI

embeddings = AzureOpenAI(
    azure_deployment="text-embedding-ada-002",
    api_version="2024-02-15-preview"
)
```

```python
collection_name = "web"
query = "Quels sont les forfaits téléphoniques français avec itinérance aux États-Unis?"

query_vector=embeddings.embeddings.create(
    input=[query],
    model="text-embedding-ada-002"
)
```

```python
print(query_vector.data)

search_result = client.search(
    collection_name=collection_name,
    query_vector=query_vector.data[0].embedding,
    limit=5
)

print(search_result)
```

```python
import json

for result in search_result:
    print(result.payload[ "payload" ])
    json_result = json.loads(result.payload[ "payload" ])

    for key in json_result:
        print(json_result[ key ])
```
